*Support Vector Machine*

The Support Vector Machine, or SVM, is a linear model that can be used to solve classification and regression issues. It can solve both linear and nonlinear problems and is useful for a wide range of applications. SVM is a basic concept: The method divides the data into classes by drawing a line or hyperplane.

```python
#load dataset
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\lenovo\Downloads\archive\Social_Network_Ads.csv")
```

```python
dataset
```

```python
#data preprocessing
#train and test
X = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4]
```

```python
X,y
```

```python
#model_selection --> train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.25,random_state=0)
```

```python
X_train, y_train
```

```python
X_test, y_test
```

```python
#scale down
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = sc.fit_transform(X_test)
```

```python
X_test
```

```python
#model creation
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
#performance - accurcy socre, classification -- confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Accuracy: ", accuracy_score(y_test,y_pred))
```

```python
cf = confusion_matrix(y_test,y_pred)
```

```python
print(cf)
```
